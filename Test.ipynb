```python
lis1=[1,2,3,4]
lis2=[2,6,7,8,9,10]

def merg(lis1,lis2):
    i=j=0
    res=[]
    while i<len(lis1) and j<len(lis2):
        if lis1[i]<lis2[j]:
            res.append(lis1[i])
            i=i+1
        else:
            res.append(lis2[j])
            j=j+1
    res.extend(lis1[i:]+lis2[j:])
    return res
merg(lis1,lis2)
```
