```python
lis1=[1,2,3,4]
lis2=[2,6,7,8,9,10]

def merg(lis1,lis2):
    i=j=0
    res=[]
    while i<len(lis1) and j<len(lis2):
        if lis1[i]<lis2[j]:
            res.append(lis1[i])
            i=i+1
        else:
            res.append(lis2[j])
            j=j+1
    res.extend(lis1[i:]+lis2[j:])
    return res
merg(lis1,lis2)
```

```python
import numpy as np
import pandas as pd

# Simulate training (expected) data — balanced, normal range
np.random.seed(42)
train_preds = np.random.beta(a=2, b=5, size=1000)  # more low probabilities (most non-pneumonia)

# Simulate production (actual) data — drifted (more pneumonia-like patterns)
prod_preds = np.random.beta(a=2, b=4, size=1000)   # more uniform distribution

# Combine into DataFrames for clarity
train_df = pd.DataFrame({'prediction_probability': train_preds})
prod_df = pd.DataFrame({'prediction_probability': prod_preds})

print("Training Data Sample:")
print(train_df.head())

print("\nProduction Data Sample:")
print(prod_df.head())
```

```python
import matplotlib.pyplot as plt

plt.hist(train_preds, bins=20, alpha=0.6, label='Training', density=True)
plt.hist(prod_preds, bins=20, alpha=0.6, label='Production', density=True)
plt.title('Prediction Probability Distribution')
plt.xlabel('P(pneumonia)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
def calculate_psi(expected, actual, bins=10):
    breakpoints = np.linspace(0, 1, bins + 1)
    expected_perc = np.histogram(expected, bins=breakpoints)[0] / len(expected)
    actual_perc = np.histogram(actual, bins=breakpoints)[0] / len(actual)
    psi = np.sum((actual_perc - expected_perc) * np.log((actual_perc + 1e-6) / (expected_perc + 1e-6)))
    return psi

psi_value = calculate_psi(train_df['prediction_probability'], prod_df['prediction_probability'])
print(f"PSI Value: {psi_value:.3f}")
```
